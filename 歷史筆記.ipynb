```python
import os
import sys
史前台灣時空架構圖 = '史前台灣時空架構圖.png'
荷治時期台灣對外貿易圖 = '荷治時期台灣對外貿易圖.png'
鄭氏時期台灣對外貿易圖 = '鄭氏時期台灣對外貿易圖.png'
while True:
    print()
    print('----------歷史講義筆記----------')
    選擇 = input('請輸入查詢單元 (U1~) or stop :')
    if 選擇 == 'stop':
        break
    elif 選擇.lower() == 'u1':
        try:
         with open('U1.txt', 'r', encoding ='utf-8') as f:
             content = f.read()
             print()
             print(content)
        except FileNotFoundError:
         print("錯誤：找不到檔案 'U1.txt'。")
        # --- 這是您需要加上的「安全網」程式碼 ---(Google Gemini)
        try:
            # 步驟 1: 將可能出錯的程式碼，放在 try 區塊裡面
            print('----------------------------------------------------')
            print()
            print(f"正在嘗試開啟圖片 '{史前台灣時空架構圖}'...")
            if sys.platform == "win32":
                os.startfile(史前台灣時空架構圖)
            else:
                print("此功能主要支援 Windows 系統。")
        # 步驟 2: 指定要捕捉的錯誤類型，並提供解決方案
        except FileNotFoundError:
            # 如果在 try 區塊中發生了 FileNotFoundError，程式會立刻跳到這裡執行
            print(f"\n【錯誤】找不到圖片檔案 '{史前台灣時空架構圖}'！")
            print("請確認：")
            print("1. 檔案名稱完全正確 (包含 .png)")
            print("2. 圖片檔案和您的 .py 程式放在同一個資料夾。\n")
        # 程式會繼續往下執行，不會因為找不到檔案而崩潰
        print("圖片處理流程已結束，程式繼續正常運作。")
    elif 選擇.lower() == 'u2':
        try:
            with open('U2.txt', 'r', encoding ='utf-8') as d:
                content = d.read()
                print()
                print(content)
        except FileNotFoundError:
          print("錯誤：找不到檔案 'U2.txt'。")
        try:
            print('----------------------------------------------------')
            print()
            print(f"正在嘗試開啟圖片'{荷治時期台灣對外貿易圖}''...'")
            print(f"正在嘗試開啟圖片'{鄭氏時期台灣對外貿易圖}''...'")
            if sys.platform == "win32":
                os.startfile(荷治時期台灣對外貿易圖)
                os.startfile(鄭氏時期台灣對外貿易圖)
            else:
                print("此功能主要支援 Windows 系統。")
        except FileNotFoundError:
            print(f"\n【錯誤】找不到圖片檔案 '{荷治時期台灣對外貿易圖}'！")
            print(f"\n【錯誤】找不到圖片檔案 '{鄭氏時期台灣對外貿易圖}'！")
            print("請確認：")
            print("1. 檔案名稱完全正確 (包含 .png)")
            print("2. 圖片檔案和您的 .py 程式放在同一個資料夾。\n")
        print("圖片處理流程已結束，程式繼續正常運作。")
```
